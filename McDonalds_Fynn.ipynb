```python
# Select the columns to modify (the first 11 columns seem to contain 'Yes'/'No' values)
cols_to_modify = df.columns[:11]

# Apply the replacement using a loop
for col in cols_to_modify:
    df[col] = df[col].replace({'Yes': 1, 'No': 0})

# Display the head of the dataframe to check the changes
display(df.head())
```

```python
df.info()
```

```python
new_df = df.drop(columns= ['Like',	'Age',	'VisitFrequency',	'Gender'])
```

```python
new_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
from sklearn.decomposition import PCA
```

```python
scaled_df = scalar.fit_transform(new_df)
```

```python
pca = PCA(n_components=11)
principal_components = pca.fit_transform(scaled_df)
pca_df = pd.DataFrame(data=principal_components, columns=["PCA1","PCA2","PCA3","PCA4","PCA5","PCA6","PCA7","PCA8","PCA9","PCA10","PCA11"])
pca_df.head()
```

```python
# visualize pca1 and pca2
sns.scatterplot(data=pca_df, x='PCA1' ,y="PCA2", palette="deep")
```

```python
sns.heatmap(new_df.corr())
```

Extracting segments

```python

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

from yellowbrick.cluster import KElbowVisualizer

```

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k =(1,8))
visualizer.fit(scaled_df)
visualizer.show()
```

```python
from yellowbrick.cluster import SilhouetteVisualizer

model = KMeans(4)
visualizer = SilhouetteVisualizer(model)

visualizer.fit(scaled_df)    # Fit the data to the visualizer
visualizer.show()
```

```python
#assign cluster number with there type in PCA
kmeans_model=KMeans(4)
kmeans_model.fit_predict(scaled_df)
pca_df_kmeans= pd.concat([pca_df,pd.DataFrame({'cluster':kmeans_model.labels_})],axis=1)
```

```python
pca_df_kmeans.head()
```

```python
plt.figure(figsize=(8,8))
ax=sns.scatterplot(x="PCA1",y="PCA2",hue="cluster",data=pca_df_kmeans,palette=['yellow','teal','blue','black'])
plt.title("Clustering using K-Means Algorithm")
plt.show()
```

```python
data_new= pd.concat([df,pd.DataFrame({'cluster':kmeans_model.labels_})],axis=1)
data_new.head()
```

```python
data_new['Gender'].unique()
```

```python
data_new["Gender"]= data_new["Gender"].replace({'Female':0, 'Male':1})
```

```python
data_new['Like'].unique()
```

```python
data_new["Like"]= data_new["Like"].replace({'I hate it!-5':0, '-4':1, '-3':2, '-2':3, '-1':4, '0':5, '+1':6, '+2':7, '+3':8, '+4':9, 'I love it!+5':10})
```

```python
data_new['VisitFrequency'].unique()
```

```python
data_new["VisitFrequency"]= data_new["VisitFrequency"].replace({'Never':0, 'Once a year':1, 'Every three months':2, 'Once a month':3, 'Once a week':4, 'More than once a week':5})
```

```python

data_new.head()
```

# selection of target segment


```python
from sklearn.model_selection import train_test_split
```

```python
#Split Dataset
X = data_new.drop(['cluster'],axis=1)
y= data_new[['cluster']]
X_train, X_test, y_train, y_test =train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
clf = DecisionTreeClassifier(random_state=1234)
model = clf.fit(X, y)
```

```python
fig = plt.figure(figsize=(25,20))
tree.plot_tree(clf,   filled=True)
```

```python
#Decision_Tree
model= DecisionTreeClassifier(criterion="entropy")
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
```

```python
#Confusion_Matrix
print(metrics.confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
